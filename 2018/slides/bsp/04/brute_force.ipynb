```python
al = "abcdefghijklmnopqrstuvwxyz"
keyspace = len(al)
```

```python
def decrypt(cyphered_text, key):
    output = ""
    for character in cyphered_text:
        character = character.lower()
        if not character in al:
            output += character
            continue
        new_index = (al.index(character) - key + keyspace) % keyspace
        output += al[new_index]
    return output
```

```python
for key in range(0,26):
    print(key)
    print(decrypt(cyphered_text,key))
```

```python
def freq_analysis(cyphered_text):
    d = {}
    for character in cyphered_text:
        if not character in al:
            continue
        d.update({character: d.setdefault(character,0)+1})
    return d
```

```python
freq_analysis(cyphered_text)
```

```python
def get_most_popular_char(al_dict):
    star = ""
    max_count = 0
    for character, count in al_dict.items():
        if count > max_count:
            max_count = count
            star = character
    return star
```

```python
get_most_popular_char(freq_analysis(cyphered_text))
```

```python
def get_key(character):
    index_e = 4
    key = (keyspace - index_e + al.index(character)) % keyspace
    return key
```

```python
freq = freq_analysis(cyphered_text)
```

```python
most_pop_char = get_most_popular_char(freq)
```

```python
key = get_key(most_pop_char)
```

```python
print(decrypt(cyphered_text, key))
```
